# Import Packages

```python
import os
import sys
import json
import csv

# numeric, scientific
from scipy import linalg as spla
import numpy as np

# plotting
%matplotlib inline
import matplotlib
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
import matplotlib.pyplot as plt

# import Kitaev Honeycomb package
import kithcmb
from kithcmb import VortexSectorThermalFermions as vs

# import jsci, CT's enhanced json stream write package
import jsci
from jsci import WriteStream as jsciwrite
from jsci import Coding as jscicoding
```

# Set system parameters and initialise 

```python
L = 20
J = 1
kappa = 0.1
```

```python
kh_sys = vs.VortexSectorThermalFermions(L,J,kappa,0.01)
kh_sys.project_fermions = 1
```

### Set vortex sector to an interpolation between NV and FV

```python
kh_sys.set_adiabatic_interpolation_NV_to_FV( -1.0 )
```

Image the J matrices

```python
plt.figure( figsize=(6,6) )
plt.imshow(kh_sys.Jx)
plt.colorbar()
plt.show()

plt.figure( figsize=(6,6) )
plt.imshow(kh_sys.Jy)
plt.colorbar()
plt.show()

plt.figure( figsize=(6,6) )
plt.imshow(kh_sys.Jz)
plt.colorbar()
plt.show()
```

Image the U matrices

```python
plt.figure( figsize=(6,6) )
plt.imshow(kh_sys.Ux)
plt.colorbar()
plt.show()

plt.figure( figsize=(6,6) )
plt.imshow(kh_sys.Uy)
plt.colorbar()
plt.show()

plt.figure( figsize=(6,6) )
plt.imshow(kh_sys.Uz)
plt.colorbar()
plt.show()
```

Image the vortices matrix

```python
plt.figure( figsize=(6,6) )
plt.imshow(kh_sys.vortices)
plt.colorbar()
plt.show()
```

Image the A matrix

```python
plt.figure( figsize=(10,10) )
plt.imshow(kh_sys.A - np.transpose(kh_sys.A))
plt.colorbar()
plt.show()
```

Plot the spectrum
