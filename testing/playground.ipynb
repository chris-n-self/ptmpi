```python
%matplotlib inline
import numpy as np
# import Kitaev Honeycomb package
import kithcmb
from kithcmb import ThermalVortexSector as vs
import matplotlib.pyplot as plot
```

```python
L = 20
J = 1
kappa = 0.1
```

```python
sys = vs.ThermalVortexSectorRealSpaceCorrelations(L,J,kappa,1.0)
```

```python
plot.figure(figsize=(10, 8))
plot.plot(sys.spectrum,'.')
plot.plot([0,800],[1,1])
```

```python
plot.figure(figsize=(10, 8))
plot.plot([ np.log(np.abs(sys.spectrum[i]-sys.spectrum[i-1])) for i in range(len(sys.spectrum)-1) ])
line = -2
plot.plot( [0,800],[line,line] )
```
